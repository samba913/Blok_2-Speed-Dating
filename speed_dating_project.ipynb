```python
#data_unique_woman = data_unique.loc[data_unique["Genre"] == "woman"]
#data_unique_man = data_unique.loc[data_unique["Genre"] == "man"]




#data_unique = data.groupby("iid").first().reset_index()
#data_unique
#data_unique["Genre"].value_counts()

unique_id_woman = unique_id[unique_id["gender"] == "Woman"]
unique_id_man = unique_id[unique_id["gender"] == "Man"]

# What women think men are looking for, compared to what men are really looking for.

# Compute means for each attribute for women and men separately
mean_attributes_women = unique_id_woman[["attr2_1", "sinc2_1", "intel2_1", "fun2_1", "amb2_1", "shar2_1"]].mean()
mean_attributes_men =unique_id_man[["attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"]].mean()

categories = ['attractive', 'sincere', 'intelligente', 'fun', 'ambitious', 'interest']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=mean_attributes_men,
    theta=categories,
    fill='toself',
    name='what men are looking for'
))

fig.add_trace(go.Scatterpolar(
    r=mean_attributes_women,
    theta=categories,
    fill='toself',
    name='what women think men are looking for'
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 40])),
    showlegend=True,
    title_text="what women think are looking for VS what men are really looking for",
    title_x=0.42
)

fig.show()
```

on vois que les femmes pensent sur estimes les attentes des hommes en termes d'attirance et sous estimes leur attentes intelectuels et d'honneteté


we see that women overestimate men's expectations in terms of attraction and underestimate their intellectual and honesty expectations

Les hommes font à peu près les mêmes erreurs que les femmes. Ils surestiment leurs attentes en termes d’attirance et sous-estiment l’importance de l’intelligence et de l’honnêteté pour les femmes.
